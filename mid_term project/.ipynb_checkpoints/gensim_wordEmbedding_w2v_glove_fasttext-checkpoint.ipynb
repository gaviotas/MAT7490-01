```python
from sklearn.manifold import TSNE
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
vocab_w2v=list(model_w2v.wv.vocab)
vocab_glove=list(model_glove.wv.vocab)
vocab_fasttext=list(model_fasttext.wv.vocab)
```

```python
X_w2v=model_w2v[vocab_w2v[0:100]]
print(X_w2v)
```

```python
print(len(X_w2v))
print(X_w2v[0][:10])
```

```python
tsne = TSNE(n_components=2)

# 100개의 단어에 대해서만 시각화
X_w2v_tsne = tsne.fit_transform(X_w2v[:100,:])
# X_tsne = tsne.fit_transform(X)
```

```python
import pandas as pd
df = pd.DataFrame(X_w2v_tsne, index=vocab_w2v[:100], columns=['x', 'y'])
df.shape
```

```python
df.head(10)
```

```python
fig = plt.figure()
fig.set_size_inches(40, 20)
ax = fig.add_subplot(1, 1, 1)

ax.scatter(df['x'], df['y'])

for word, pos in df.iterrows():
    ax.annotate(word, pos, fontsize=30)
plt.show()
```
